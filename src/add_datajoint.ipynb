```python
data = pd.DataFrame((SessionStats() ).fetch())
mice_probe = {}
for m in np.unique(data.subject_name):
    for probe in np.unique(data[data.subject_name == m].probe_num):
        dates = data[(data.subject_name == m) & (data.probe_num == probe)].session_datetime
        if len(dates)>30: # mice recorded for over 30 days
            mice_probe[f'{m}_{probe}']=dict(days = (dates - np.min(dates))// np.timedelta64(1, 'D'),
                                            sua = data[(data.subject_name == m) & (data.probe_num == probe)].n_sua_units)

%matplotlib widget
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
fig = plt.figure(figsize = [8,3])
ax = fig.add_axes([0.15,0.15,0.5,0.8])
ax.spines[['right', 'top']].set_visible(False)

for d in mice_probe.keys():
    dset = mice_probe[d]
    plt.plot(dset['days'],dset['sua'],'o',alpha = 0.5,label = d)
    # plt.plot(data['days'],data['n_mua_units'],'ro')
plt.legend()
plt.gca().set_yscale('linear')
plt.ylim(10,300)
plt.ylabel('Simultaneously recorded \ningle units per probe',fontsize = 14)
plt.xlabel('Days since first recording',fontsize = 14)

ax = fig.add_axes([0.75,0.15,0.2,0.8],sharey=ax)
ax.spines[['right', 'top']].set_visible(False)

averages = []
for d in mice_probe.keys():
    dset = mice_probe[d]
    idx1 = dset['days']<15
    idx2 = (dset['days']>60) & (dset['days']>75) 
    averages.append([np.mean(dset['sua'][idx1]),np.mean(dset['sua'][idx2])])
    plt.plot(averages[-1],'o-',alpha = 0.5)
plt.xticks([0,1],['2 weeks', '2 months'])
```
